```python
sns.barplot(
    data=df,
    x="Student",
    y="Python_Marks",
    hue="Student",
    palette="viridis",
    legend=False,
)
plt.title("Python Marks of Students")
plt.xlabel("Student Name")
plt.ylabel("Python Marks")
plt.show()
```

```python
sns.barplot(
    data=df,
    x="Student",
    y="Python_Marks",
    hue="Student",
    palette="viridis",
    legend=False,
)
plt.title("Python Marks of Students")
plt.xlabel("Student Name")
plt.ylabel("Python Marks")
plt.xticks(rotation=45) ## the fhange from above
plt.show()
```

```python
sns.barplot(
data=df,
x="Department",
y="Python_Marks",
hue="Department",
    palette="plasma",
    legend=False,
)
plt.title("Average Python Marks by Department")
plt.xlabel("Department")
plt.ylabel("Average Python Marks")
plt.show()
```

```python
sns.countplot(
data=df,
x="Department",
hue="Department",
    palette="Set2",

)
plt.title("Number of Students in Each Department")
plt.xlabel("Department")
plt.ylabel("Number of Students")
plt.show()
```

```python
sns.histplot(
data=df,
x="Python_Marks",
hue="Department",
palette="coolwarm",
)
plt.title("Distribution of Python Marks")
plt.xlabel("Python Marks")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.histplot(
data=df,
x="Python_Marks",
kde=True
)
plt.title("Distribution of Python Marks")
plt.xlabel("Python Marks")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.histplot(
data=df,
x="Python_Marks",
bins=5, # means the data is divided into five intervals.
kde=True
)
plt.title("Distribution of Python Marks")
plt.show()
```

```python
sns.boxplot(
data=df,
y="Python_Marks"
)
plt.title("Box Plot of Python Marks")
plt.ylabel("Python Marks")
plt.show()
```

```python
sns.boxplot(
data=df,
x="Department",
y="Python_Marks"
)
plt.title("Python Marks by Department")
plt.xlabel("Department")
plt.ylabel("Python Marks")
plt.show()
```

```python
sns.scatterplot(
data=df,
x="Attendance",
y="Python_Marks"
)
plt.title("Attendance versus Python Marks")
plt.xlabel("Attendance")
plt.ylabel("Python Marks")
plt.show()
```

```python
sns.scatterplot(
data=df,
x="Attendance",
y="Python_Marks",
hue="Department"
)
plt.title("Attendance versus Python Marks")
plt.xlabel("Attendance")
plt.ylabel("Python Marks")
plt.show()
```

```python
sns.scatterplot(
data=df,
x="Attendance",
y="Python_Marks",
hue="Department",
size="Average_Marks"
)
plt.title("Attendance, Python Marks and Average Marks")
plt.show()
```

```python
df["Student_Number"] = range(1, len(df) + 1)
```

```python
sns.lineplot(
data=df,
x="Student_Number",
y="Average_Marks",
marker="o"
)
plt.title("Average Marks by Student Number")
plt.xlabel("Student Number")
plt.ylabel("Average Marks")
plt.show()
```
