```python
array = [3, 4, 5, 12, 12, 8, 9]

a = array[0]
positionoflargest = 0
positionoflargest1 = 0
for i in range(len(array)):
    if array[i] > a:
        a = array[i]
        positionoflargest = i
    if array[i] == a:
        positionoflargest1 = i

print(a)
print(positionoflargest)
print(positionoflargest1)
```
