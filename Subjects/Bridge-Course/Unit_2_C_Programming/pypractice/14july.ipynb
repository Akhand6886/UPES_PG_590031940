```python
words = ['abc', 'bcd', 'b', 'cde', 'abcde']


for i in range(len(words)):

    swapped = False
    

    for j in range(0, len(words) - i - 1):

        if words[j] > words[j + 1]:

            words[j], words[j + 1] = words[j + 1], words[j]
            swapped = True
            

    if not swapped:
        break

print(words)
```
